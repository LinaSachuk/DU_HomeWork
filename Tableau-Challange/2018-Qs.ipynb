```python
# Import CSV files for 12 months of 2018

df_1 = pd.read_csv("2018/JC-201801-citibike-tripdata.csv")
df_2 = pd.read_csv("2018/JC-201802-citibike-tripdata.csv")
df_3 = pd.read_csv("2018/JC-201803-citibike-tripdata.csv")
df_4 = pd.read_csv("2018/JC-201804-citibike-tripdata.csv")
df_5 = pd.read_csv("2018/JC-201805-citibike-tripdata.csv")
df_6 = pd.read_csv("2018/JC-201806-citibike-tripdata.csv")
df_7 = pd.read_csv("2018/JC-201807-citibike-tripdata.csv")
df_8 = pd.read_csv("2018/JC-201808-citibike-tripdata.csv")
df_9 = pd.read_csv("2018/JC-201809-citibike-tripdata.csv")
df_10 = pd.read_csv("2018/JC-201810-citibike-tripdata.csv")
df_11 = pd.read_csv("2018/JC-201811-citibike-tripdata.csv")
df_12 = pd.read_csv("2018/JC-201812-citibike-tripdata.csv")
```

```python
df_12.head()
```

```python
# Create a df for a year

year_2018 = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12], join = "outer")
year_2018.head(20)
```

```python
# Create Trip Minutes column 
year_2018["Trip Minutes"] = round(year_2018["tripduration"]/ 60, 2)
year_2018.head()
```

```python
year_2018.loc[year_2018['gender'] == 1, 'gender'] = 'male' 
year_2018.loc[year_2018['gender'] == 2, 'gender'] = 'female' 
year_2018.loc[year_2018['gender'] == 0, 'gender'] = 'unknown' 
year_2018.head()
```

```python

# to CSV 

year_2018.to_csv("Resources/2018.csv", index = False)
```
