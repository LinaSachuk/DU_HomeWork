## Purchasing Analysis (Total)

```python
# number of unique items
unique_items = df.groupby('Item ID')['Item ID'].nunique()
count_it = unique_items.count()
count_it
```

```python
summary_df['Average Price'] = summary_df['Average Price'].map("${: .2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:,}".format)

summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
df = pd.read_csv(file_to_load)
df.head()
```

```python
# Establish bins for ages
print(df["Age"].max())
print(df["Age"].min())
```

```python
# Create bins 
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99,
        34.99, 39.99, 99.99]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]
```

```python
# Slice the data and place it into bins
# Categorize the existing players using the age bins. Hint: use pd.cut()
pd.cut(df["Age"], bins, labels=group_labels).head()
```

```python
df["Age Group"] = pd.cut(df["Age"], bins, labels=group_labels)
df.head()
```

```python
# Create new data frame with the added "Age Group" and group it
age_groupby = df.groupby("Age Group")

# Count total players by age category
total_count_age = age_groupby["SN"].nunique()
# print(total_count_age)

# Calculate percentages by age category 
percentage_by_age = (total_count_age/total_players) * 100
# print(percentage_by_age)

# Create data frame 
age_demographics = pd.DataFrame({"Total Count": total_count_age, 
                                 "Percentage of Players": percentage_by_age})
# age_demographics
# Format the data frame with no index name in the corner
age_demographics.index.name = None

# Format percentage with two decimal places 
age_demographics['Percentage of Players'] = age_demographics["Percentage of Players"].map("{: .2f}%".format)
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data data frame by age
df = pd.read_csv(file_to_load)
df.head()

```

```python
# Establish bins for ages
print(df["Age"].max())
print(df["Age"].min())

```

```python
# Create bins 
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99,
        34.99, 39.99, 99.99]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

```

```python
# Slice the data and place it into bins
# Categorize the existing players using the age bins. Hint: use pd.cut()
pd.cut(df["Age"], bins, labels=group_labels).head()

```

```python
df["Age Group"] = pd.cut(df["Age"], bins, labels=group_labels)
df.head()
```

```python
df_groupby = df.groupby('Age Group')
purchase_count = df_groupby['Purchase ID'].count()
purchase_count
```

```python
# Obtain average purchase price by age group 
avg_purchase_price = df_groupby["Price"].mean()
avg_purchase_price
```

```python
# Calculate total purchase value by age group 
total_purchase_value = df_groupby["Price"].sum()
total_purchase_value
```

```python
total_count = df_groupby["SN"].nunique()
```

```python
# Calculate the average purchase per person in the age group 
avg_purchase_per_person = total_purchase_value/total_count
avg_purchase_per_person
```

```python
Purchasing_Analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": avg_purchase_price,
                                   "Total Purchase Value": total_purchase_value,
                                   "Avg Total Purchase per Person": avg_purchase_per_person})
Purchasing_Analysis
```

```python
Purchasing_Analysis.index.name = None
         
Purchasing_Analysis
```

```python
Purchasing_Analysis["Average Purchase Price"] = Purchasing_Analysis["Average Purchase Price"].map("${:,.2f}".format)
Purchasing_Analysis["Total Purchase Value"] = Purchasing_Analysis["Total Purchase Value"].map("${:,.2f}".format)
Purchasing_Analysis["Avg Total Purchase per Person"] = Purchasing_Analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)

Purchasing_Analysis
```
